```python
import aiohttp
import asyncio 

pflask = 8766
psanic = 8765
async def requestDb(port):
    async with aiohttp.ClientSession() as session:
        async with session.get(f'http://127.0.0.1:{port}/rundb') as resp:
            print(resp.status)
            print(await resp.text())

task = asyncio.create_task(requestDb(pflask))
```

```python
tasks = [asyncio.create_task(requestDb(psanic)) for _ in range(100)]
```

```python
tasks = [asyncio.create_task(requestDb(pflask)) for _ in range(10)]
```
